# ebay Motorcycle Listing

### The data used in this final project is data obtained from the marketplace site, ebay.com. The data used are motorcycle for predict price of Motorcycle

### Name of Features 
Condition = Condition

Condition_Desc = Condition Description

Price = Price

Location = Location

Model Year = Model Year

Mileage = Mileage on Vehicle.

Exterior_Color = Exterior Color.

Make = Make/Brand

Warranty = Does the vehicle have a warranty?

Model = Model of Bike

Sub_Model = Optional Sub Model Field.

Type = Type of Motorcycle

Vehicle_Title = Vehicle Title

OBO = Was there an option to offer a different amount than the listed price?

Feedback_Perc = Percentage of Positive Feedback the Seller has received. 

Watch_Count = Number of people watching item at time the data was scraped.

N_Reviews = Number of reviews for the ebay seller account attached to said listing.

Auction = The item for Auction or no

Buy_Now = The Auction Buy now or no

### Cleaning process column 'Make'

```python
df.Make.unique()
```

```python
df.drop(df[df['Make'].isnull()].index, inplace = True)
```

```python
def apply_Make(x):
    x = str(x)
    if('harley-davidson' in x.lower()):
        return 'Harley-Davidson'
    elif('yamaha' in x.lower()) :
        return 'Yamaha'
    elif('kawasaki' in x.lower()):
        return 'Kawasaki'
    elif('suzuki' in x.lower()) :
        return 'Suzuki'
    elif('bmw' in x.lower()):
        return 'BMW'
    elif('triumph' in x.lower()) :
        return 'Triumph'
    elif('ducati' in x.lower()):
        return 'Ducati'
    elif('custom' in x.lower()) :
        return 'custom'
    elif('can-am' in x.lower()):
        return 'can-am'
    elif('polaris' in x.lower()):
        return 'Polaris'
    elif('honda' in x.lower()):
        return 'Honda'
    elif('KTM' in x.lower()):
        return 'KTM'
    elif('ural' in x.lower()):
        return 'Ural'
    elif('moto guzzi' in x.lower()):
        return 'Moto Guzzi'
    elif('american' in x.lower()):
        return 'American Motors'
    elif('aprilia' in x.lower()):
        return 'Aprilia'
    elif('buell' in x.lower()):
        return 'Buell'
    elif('Royal Enfield' in x.lower()):
        return 'Royal Enfield'
    else:
        return 'Other Makes'
    
df['Make'] = df['Make'].apply(apply_Make)
```

### Cleaning process column 'Seller_Status'

```python
df.Seller_Status.unique()
```

```python
def seller_status(x):
    x = str(x)
    if('owner' in x.lower()):
        return 'Owner'
    elif('dealer' in x.lower()) :
        return 'Dealer'
    elif('private seller' in x.lower()):
        return 'Owner'
    else:
        return 'Company'
    
df.Seller_Status = df.Seller_Status.apply(seller_status)
```

```python
df.Seller_Status.value_counts()
```

### Cleaning process column Warranty

```python
df.Warranty.unique()
```

```python
df.Warranty.value_counts().head()
```

```python
def warranty(x):
    x = str(x)
    if('no' in x.lower()):
        return 'No'
    elif('Unspecified' == x) :
        return 'Unknown'
    else:
        return 'Yes'
    
df['Warranty'] = df.Warranty.apply(warranty)
```

```python
df.Warranty.value_counts()
```

### Cleaning process column 'Feedback_perc'

```python
df.Feedback_Perc.value_counts()
```

```python
df['Feedback_Perc'].fillna(0,inplace = True)
```

```python
def delete_lebih_dari(item):
    if '>' in str(item):
        item = (str(item).split('>')[1])
        return round(float(item))
    else :
        return round(float(item))
    
df['Feedback_Perc'] = df['Feedback_Perc'].apply(delete_lebih_dari)
```

### Cleaning process column Vehicle_Title

```python
def fn_apply_vehicle_title(col):
    if(pd.isna(col['Vehicle_Tile'])):
        return df[df['Make'] == col['Make']]['Vehicle_Tile'].mode()[0]
    else:
        return col['Vehicle_Tile']
```

```python
df['Vehicle_Tile'] = df[['Vehicle_Tile','Make']].apply(fn_apply_vehicle_title,axis=1)
```

```python
df = df.rename(columns={'Vehicle_Tile': 'Vehicle_Title'})
```

```python
df.Vehicle_Title.value_counts()
```

```python
def rebuilt(x):
    if('Rebuilt, Rebuildable &amp; Reconstructed' == x): 
        return 'Rebuilt & Reconstructed'
    else:
        return x

df['Vehicle_Title'] = df.Vehicle_Title.apply(rebuilt)
```

### Cleaning process column 'Type'

```python
df.Type.unique()
```

```python
def apply_Type(x):
    x = str(x)
    if('gasoline' in x.lower()):
        return 'gasoline'
    elif('cruiser' in x.lower()) :
        return 'cruiser'
    elif('standard' in x.lower()):
        return 'standard'
    elif('touring' in x.lower()) :
        return 'touring'
    elif('dual-purpose' in x.lower()):
        return 'dual purpose'
    elif('racer' in x.lower()) :
        return 'Racer'
    elif('super' in x.lower()):
        return 'super'
    elif('softail' in x.lower()):
        return 'softail'
    elif('scooter' in x.lower()):
        return 'scooter'
    elif('sport' in x.lower()):
        return 'sport'
    elif('chopper' in x.lower()):
        return 'chopper'
    elif('off-road' in x.lower()):
        return 'Off Road'
    elif('classic' in x.lower()):
        return 'classic'
    elif('custom' in x.lower()):
        return 'custom'
    elif('naked' in x.lower()):
        return 'naked'
    else:
        return 'Others'
    
df['Type']= df['Type'].apply(apply_Type)
```

### Cleaning process column Mileage

```python
df.Mileage.head(10)
```

```python
def replace_comma(item):
    if(type(item) != float):
        if(',' in item):
            item = item.replace(',','.')
            return item
        else:
            return item
    return item
df['Mileage']= df['Mileage'].apply(replace_comma)
```

```python
def ubah_int(item):
    try:
        return float(item)
    except:
        return None
```

```python
df['Mileage'] = df['Mileage'].apply(ubah_int)
```

```python
df.drop(df[df['Mileage'].isnull()].index,inplace=True)
```

### Cleaning process column Model_Year

```python
def clean_year(item):
    return int(item)
        
df['Model_Year'] = df['Model_Year'].apply(clean_year)
```

### Cleaning process Column Price

```python
def clean_price(item):
    item = str(item)[1:].replace(',','')
    return int(item)

df['Price'] = df['Price'].apply(clean_price)
```

### Cleaning process column 'Exterior_Color'

```python
def apply_color(x):
    x = str(x)
    if('black' in x.lower()):
        return 'black'
    elif('red' in x.lower()) :
        return 'red'
    elif('blue' in x.lower()):
        return 'blue'
    elif('white' in x.lower()) :
        return 'white'
    elif('silver' in x.lower()):
        return 'silver'
    elif('orange' in x.lower()) :
        return 'orange'
    elif('gray' in x.lower()):
        return 'gray'
    elif('green' in x.lower()) :
        return 'green'
    elif('burgundy' in x.lower()):
        return 'burgundy'
    elif('yellow' in x.lower()):
        return 'yellow'
    elif('gold' in x.lower()):
        return 'gold'
    elif('purple' in x.lower()):
        return 'purple'
    elif('custom' in x.lower()):
        return 'custom'
    else:
        return 'Other Colors'
    
df['Exterior_Color'] = df['Exterior_Color'].apply(apply_color)
```

### Check Outlier column Price and Mileage

```python
def check_outlier(dataframe, col):
    q1 = dataframe[col].quantile(0.25)
    q3 = dataframe[col].quantile(0.75)
    iqr = q3-q1
    upper_limit = q3 + (1.5 * iqr)
    lower_limit = q1 - (1.5 * iqr)
    return dataframe[(dataframe[col] < lower_limit) | (dataframe[col] > upper_limit)].index ,upper_limit,lower_limit
```

```python
id_to_drop,upper,lower = check_outlier(df,'Price')
```

```python
df.drop(id_to_drop, inplace =True)
```

```python
id_to_drop_mileage,upper,lower = check_outlier(df,'Mileage')
```

```python
df.drop(id_to_drop_mileage,inplace= True)
```
